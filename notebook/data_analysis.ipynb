```python
import scipy.stats as stats
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define the file path
file_path = ("..\data\MachineLearningRating_v3.txt")

# Read the data into a DataFrame
df = pd.read_csv(file_path, delimiter='|')

# Define the file path for the output CSV file
output_file_path = ("..\data\MLearningRating_v3.csv")

# Save the DataFrame to a CSV file
df.to_csv(output_file_path, index=False)

print(f'Data saved to {output_file_path}')

```

```python
df.isnull().sum()
```

**Drop columns with more than 90% missing values **

```python
threshold = len(df) * 0.9 
df = df.dropna(thresh=threshold, axis=1)
```

**Fill missing values** 

```python
# Replace comma with dot and convert to float 
df['CapitalOutstanding'] = df['CapitalOutstanding'].str.replace(',', '.').astype(float)
df['AccountType'] = df['AccountType'].fillna(df['AccountType'].mode()[0]) 
df['MaritalStatus'] = df['MaritalStatus'].fillna(df['MaritalStatus'].mode()[0]) 
df['Gender'] = df['Gender'].fillna(method='ffill') 
df['CapitalOutstanding'] = df['CapitalOutstanding'].fillna(df['CapitalOutstanding'].median())
```

```python
# Convert the entire column to strings
df['CapitalOutstanding'] = df['CapitalOutstanding'].astype(str)

# Replace comma with dot and handle non-numeric entries gracefully
df['CapitalOutstanding'] = df['CapitalOutstanding'].str.replace(',', '.')

# Convert strings to float, coerce errors to NaN
df['CapitalOutstanding'] = pd.to_numeric(df['CapitalOutstanding'], errors='coerce')

# Fill missing values with the median
df['CapitalOutstanding'] = df['CapitalOutstanding'].fillna(df['CapitalOutstanding'].median())

# Verify and correct data types 
df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth']) 
df['IsVATRegistered'] = df['IsVATRegistered'].astype('bool')

# For categorical columns 
df['VehicleType'] = df['VehicleType'].fillna('Unknown') 
df['make'] = df['make'].fillna('Unknown') 
df['Model'] = df['Model'].fillna('Unknown') 
df['bodytype'] = df['bodytype'].fillna('Unknown') 

# For numerical columns, using median 
df['mmcode'] = df['mmcode'].fillna(df['mmcode'].median()) 
df['Cylinders'] = df['Cylinders'].fillna(df['Cylinders'].median()) 
df['cubiccapacity'] = df['cubiccapacity'].fillna(df['cubiccapacity'].median()) 
df['kilowatts'] = df['kilowatts'].fillna(df['kilowatts'].median()) 
df['NumberOfDoors'] = df['NumberOfDoors'].fillna(df['NumberOfDoors'].median()) 

# For dates, you can use the mode or a specific date 
df['VehicleIntroDate'] = pd.to_datetime(df['VehicleIntroDate'], errors='coerce') 
df['VehicleIntroDate'] = df['VehicleIntroDate'].fillna(df['VehicleIntroDate'].mode()[0])

df.info()
```

**Exploratory Data Analysis (EDA)**

```python
# Histogram of TotalPremium
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPremium'], bins=30, kde=True)
plt.title('Distribution of Total Premium')
plt.xlabel('Total Premium')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
# Histogram of TotalPremium
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalClaims'], bins=30, kde=True)
plt.title('Distribution of Total Claim')
plt.xlabel('Total Claim')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df[['TotalPremium', 'TotalClaims', 'PostalCode']].corr()

# Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='TotalPremium', y='TotalClaims', hue='PostalCode', alpha=0.6)
plt.title('Total Premium vs Total Claims by Zip Code')
plt.xlabel('Total Premium')
plt.ylabel('Total Claims')
plt.legend(title='Postal Code', bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Plot histograms for numerical columns 
numerical_columns = ['TotalPremium', 'TotalClaims', 'CapitalOutstanding'] 
df[numerical_columns].hist(bins=30, figsize=(10, 7)) 
plt.tight_layout() 
plt.show()
```

```python
# Plot bar charts for categorical columns 
categorical_columns = ['Gender', 'MaritalStatus', 'VehicleType'] 
for col in categorical_columns: 
    plt.figure(figsize=(10, 5)) 
    sns.countplot(data=df, x=col) 
    plt.title(f'Distribution of {col}') 
    plt.show()
```

**Bivariate or Multivariate Analysis**

```python
# Scatter plot and correlation matrix
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='TotalPremium', y='TotalClaims', hue='Province')
plt.title('TotalPremium vs TotalClaims by Province')
plt.show()

# Correlation matrix
correlation_matrix = df[['TotalPremium', 'TotalClaims', 'CapitalOutstanding']].corr()
plt.figure(figsize=(10, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

**Data Comparison**

```python
plt.figure(figsize=(10, 5)) 
sns.boxplot(data=df, x='Gender', y='TotalPremium') 
plt.title('TotalPremium Distribution by Gender') 
plt.show()
```

```python
df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth'])
df['Month'] = df['TransactionMonth'].dt.to_period('M')
monthly_data = df.groupby('Month')[['TotalPremium', 'TotalClaims']].sum().reset_index()
monthly_data['Month'] = monthly_data['Month'].astype(str)

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_data, x='Month', y='TotalPremium', label='Total Premium')
sns.lineplot(data=monthly_data, x='Month', y='TotalClaims', label='Total Claims')
plt.title('Monthly Total Premium and Total Claims')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
